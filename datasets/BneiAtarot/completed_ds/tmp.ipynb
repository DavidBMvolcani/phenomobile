```python
import pandas as pd
import os

import datetime
current_datetime = datetime.datetime.now()
ts = current_datetime.timestamp()

cd=os.getcwd()
files=os.listdir(f'{cd}/datasets')
files
```

```python
old_files=files
files=[]
files.append(old_files[0])
files.append(old_files[2])
files.append(old_files[3])
files.append(old_files[4])
files
```

```python
for f in files:
    df=pd.read_csv(f'{cd}/datasets/{f}')
    print(df.shape)
```

```python

# Read and concatenate
df = pd.concat((pd.read_csv(f'{cd}/datasets/{f}') for f in files), ignore_index=True)

# Save to a new CSV (optional)
df.to_csv(f"complete_HS_ds_of_Anthocyanin_BENI_ATAROT_2025_12_ts_{ts}.csv", index=False)

print("CSV files concatenated successfully!")
```
