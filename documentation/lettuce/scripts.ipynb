###### run the model on linear regression with different filters

```python
#outputs_folder=f'{os.getcwd()}/outputs'
#ds_name='complete_ds_of_Anthocyanin_BENI_ATAROT_2025-12-05_1766935305.461822.csv'

ml=training.training(dataset_name=ds_name,task='regression',model='linear regression')
print(ml.df.shape)

trainable_features=['valPlantMean','satPlantMean','huePlantMean']
trainable_features.append('Anthocyanin')
target='Anthocyanin'

#evaluate the model without condition
res_df=ml.evaluate_regression_models(trainable_features,target)

#now evaluate the model on different condtions
conds=['White and Blue Led','White Led','Shade','Control']
for c in conds:
    cur_res_df=ml.evaluate_regression_models(trainable_features,target,
                                  filter_df=True,filter_cond=c,filter_indicator='catalog id')
    res_df=pd.concat([res_df, cur_res_df])
    
res_df  
```
